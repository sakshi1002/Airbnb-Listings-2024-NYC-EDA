### EDA

Task 5: Data Analysis

**Univariate Analysis**


```python
#Identifying Outliers in price

df = df[df['price'] < 1500]

sns.boxplot(data=df, x='price')
```

```python
#Price Distribution

df['price'].min()
df['price'].max()
df['price'].mean()

plt.figure(figsize=(8,5))
sns.histplot(data=df, x='price', bins=100)
plt.ylabel('Frequency')
plt.xlabel('Price')
plt.title('Price Frequency Of Airbnb Rooms')
```

```python
#Availability Distribution
# df['availability_365']

plt.figure(figsize=(6,3))
sns.histplot(data=df, x='availability_365')
plt.ylabel('Frequency')
plt.xlabel('availability_365')
plt.title('availability Distribution')
```

```python
#neighbourhood and price 

nhood_price = df.groupby(by='neighbourhood_group')['price'].mean()
display(nhood_price)
#sns.barplot(data=nhood_price, x=df['price'], y=df['neighbourhood_group'])

```

Feature Engineering

```python
# df['price per bed']

df['price per bed'] = df['price']/df['beds']

df.head()
```

```python
# averate price per bed

nhood_bed_price = df.groupby(by='neighbourhood_group')['price per bed'].mean()

nhood_bed_price
```

Bivariate Analysis

One Variable dependency in another variable 

```python
df.columns
```

```python
#Price dependency on neighbourhood
sns.barplot(data=df, x='neighbourhood_group', y='price', hue='room_type')
plt.show()
```

```python
#no of review dependency on price

plt.figure(figsize=(10,8))

plt.title('Locality and Review Dependency')
sns.scatterplot(data=df,x='price',y='number_of_reviews', hue='neighbourhood_group')

plt.show()
```

```python
sns.pairplot(data=df, vars=['price','minimum_nights','number_of_reviews','availability_365'], hue='room_type')
```

```python
#Geographical Distribution of AirBnb Listing

sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type')
plt.title('Geographical Distribution of AirBnb Listing')
plt.show()
```
